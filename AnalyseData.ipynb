Change Data type of date and time

```python
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
```

```python
df.drop(['agent','company'],axis=1,inplace=True)
df.dropna(inplace=True)
```

Remove outlier

```python
df=df[df['adr']<5000]
```

Data Analysis And Visualizations

```python
cancelled_per=df['is_canceled'].value_counts(normalize=True)*100
print(cancelled_per)

# Create bar chart
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not cancelled','Cancelled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
plt.Figure(figsize=(8,4))
axl=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_level,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotel',size=20)
plt.xlabel('Hotel')
plt.ylabel('number of reservation')
plt.legend(['Not cancelled','Cancelled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in city and Resort Hotel',fontsize=20)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['months']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
axl=sns.countplot(x='months',hue='is_canceled',data=df,palette='bright')
legend_level=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status per Month',size=20)
plt.xlabel('months')
plt.ylabel('Number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(13,6))
plt.title('ADR per months',size=20)
sns.barplot(x='months',y='adr',data=df[df['is_canceled']==1].groupby('months')[['adr']].sum().reset_index(),palette='bright')
plt.show()
```

```python
canceled_data=df[df['is_canceled']==1]
top_10_count=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(6,6))
plt.title('Top 10 Countries with reservation canceled')
plt.pie(top_10_count,autopct='%.2f',labels=top_10_count.index)
plt.show()
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
cancelled_df_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,8))
plt.title('Average Daily rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'])
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'])
plt.legend(['Not Cancelled','Cancelled'])
plt.show()
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```
